# Cross Validation

How to find best perform algorithm for a Data

image.png

For that have to apply all algorithms for the same dataset and check the accurecy of that

```python
import pandas as pd
data = pd.read_csv('Iris.csv')
data = data.drop("Id", axis =1)
```

```python
x = data.drop("Species",axis=1)
x.head(5)
```

```python
y = data["Species"]
y.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.3)
```

## K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

## Navive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
nb.score(x_test,y_test)
```

## Support Vector Machine

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

When you try this time to time the accuracy got changed, Because we use test to split and that split the data in random manner. In that we have to use Cross - Validation

## How Implement this

```python
import pandas as pd
data = pd.read_csv("sample_data/Iris.csv")
data = data.drop("Id", axis =1)
```

```python
x = data.drop("Species", axis=1)
x.head(5)
data.head()
```

```python
y = data["Species"]
y.head(5)
```

```python
# In this use cross validation score
from sklearn.model_selection import cross_val_score
```

### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
cross_val_score(knn, x, y, cv=5).mean()
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
cross_val_score(rf, x, y, cv=5).mean()
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
cross_val_score(nb, x, y, cv=5).mean()
```

### Support Vector Machine

```python
from sklearn.svm import SVC
svm = SVC()
cross_val_score(svm, x, y, cv=5).mean()
```
