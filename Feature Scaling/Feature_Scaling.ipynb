```python
# Set data as numpy array
import numpy as np
data = np.array([[26, 50000],
                 [29, 70000],
                 [34, 55000],
                 [31, 41000]
])
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
scaled_data
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_data 
```
